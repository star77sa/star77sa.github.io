# 신호 및 시스템

`-` 신호

물리적인 현상의 본질에 대한 정보를 담고 있는 하나 이상의 변수에 관한 함수로 정의된다. 하나의 변수에 의존하면 1차원적, 함수가 두 개 혹은 그 이상의 변수에 따라 달라지면 다차원적 이라고 한다. 이미지는 가로 좌표와 세로 좌표에 따라 변하는 2차원 신호의 한 예.

`-` 시스템

임의의 기능을 수행하기 위하여 하나 이상의 신호를 처리하여 새로운 신호를 생성하는 실체(entity)

시스템 응용 - 통신 시스템, 제어 시스템, 마이크로 전자기계 시스템, 원격 센싱, 생체 신호처리, 청각 시스템

`-` 통신 시스템

모든 통신 시스템(communication system)에는 기본적으로 세 가지 구성 요소, 즉 송신기, 채널, 수신기가 존재한다. 송신기는 공간의 어느 한 점에 위치하고, 수신기는 송신기로부터 떨어진 다른 지점에 위치하며, 채널은 그 둘을 연결하는 물리적 매체

송신 신호가 채널을 통해서 전파되어감에 따라, 채널의 물리적인 특성 때문에 신호는 왜곡된다. 더욱이 잡음과 다른 원천으로부터 온 간섭 신호가 채널 출력을 변질시켜 수신 신호는 송신 신호의 왜곡된 모습의 결과로 나타난다. 수신기의 기능은, 수신 신호로부터 원래의 메시지 신호를 인식할 수 있도록 재구성하여(즉, 추정치를 만들어) 사용자에게 전달하는 것

통신 시스템은 아날로그 또는 디지털 형태로 구분된다.

신호처리 입장에서 보면 아날로그 시스템의 설계는 상대적으로 간단하다. 송신기는 변조기(modulator), 수신기는 복조기(demodulator)로 구성되어 있다.

변조(modulation)는 메시지 신호를 채널의 전송 특성에 적합하도록 적절한 형태로 변환시키는 과정이다.

보통의 경우에는, 송신 신호는 정현바송파의 진폭, 위상, 주파수의 변형된 모습으로 표현된다. 그러므로, 이들 각각을 진폭 변조, 위상 변조, 주파수 변조라 부른다. 수신기에는 이와 대응되는 진폭 복조, 위상 복조, 주파수 복조를 사용해서 원래의 메시지 신호를 추정하여 출력한다.

디지털 통신 시스템은 훨씬 더 복잡하다. 메시지 신호가 음성 신호나 영상 신호와 같은 아날로그 형태이면, 송신기는 신호를 디지털 형태로 변환하기 위하여 다음과 같은 작업을 한다.

- 샘플링(sampling): 메시지 신호를 특정 순간에서 그 진폭을 나타내는 일련의 수로 변환


- 양자화(quantization): 샘플링에 의해 생성된 각각의 수를 유한 개의 이산 진폭 레벨 가운데 가장 가까운 레벨로 표현. 예를 들어 16비트 이진수로 나타내면, 각 샘플을 216개의 레벨로 표현할 수 있다. 메시지 신호가 샘플링과 양자화 과정을 모두 거치면, 시간축과 진폭출 모두 이산적으로 표현된다.


- 코딩(coding): 양자화된 샘플을 유한 개의 심벌로 이루어진 코드 워드(code word)로 표현. 예를 들면, 이진 코드의 경우 심벌은 '1' 또는 '0' 이다.


샘플링, 코딩과는 달리 양자화는 역변환이 가능하지 않다.

즉, 양자화를 하게 되면 응용에 있어 항상 정보의 손실이 발생

만일 디지털 컴퓨터의 경우와 같이 원래의 정보가 처음부터 이산적이라면, 위에 언급된 진행 과정은 모두 필요 없게 된다.

송신기는 데이터 압축, 채널 인코딩 등과 같은 부가적인 연산을 필요로 할 수 있다.

데이터 압축의 목적은 메시지 신호로부터 중복성 정보를 제거하고, 전송에 필요한 샘플당 비트의 수를 감소시켜 채널의 효율을 높이기 위한 것

채널 인코딩은 채널을 통한 전송 과정에서 얻게 되는 잡음 및 간섭 신호 등으로부터 정보를 보호하기 위해 정해진 방식에 따라 코드 워드에 여분의 요소(예를 들면, 부가 심벌)를 추가한다.

마지막으로, 코드화된 신호는 채널을 통한 전송을 위해 반송파(대개는 정현파)로 변조된다.

수신기에서는 코딩 및 샘플링 과정이 반대로 이루어지며(즉, 각 입력과 출력 신호의 역할이 바뀌게 됨), 원래 메시지 신호의 추정치를 얻은 다음 이를 의도하는 목적지로 전달한다. 그러나 앞에서 언급한 대로 양자화는 역변환이 가능하지 않으므로, 수신기에는 이와 대응되는 것이 없다.

반도체 산업의 지속적인 발달로, 디지털 통신은 아날로그 통신보다 보통 비용면에서 더 효과적이다.

통신에는 두 가지 기본 모드가 있다.

1. 방송(broadcasting)

2. 점대점(point-to-point) 통신.

`-` LTI(Linear Time Invariant) System

- Linear = Scaling + additivity = superposition

    Scaling: system에 $x_1(t)$를 넣으면 $y_1(t)$가 나올 때, $6\times x_1(t)$를 넣으면 $6\times y_1(t)$가 나온다? Scaling을 만족.

    additivity: system에 $x_1(t)$를 넣으면 $y_1(t)$가 나오고 $x_2(t)$를 넣으면 $y_2(t)$가 나올 때 $x_3(t) = x_1(t) + x_2(t)$를 넣으면 $y_3(t) = y_1(t) + y_2(t)$가 나온다? additivity를 만족.

    superposition: $\alpha x_1(t) + \beta x_2(t) = \alpha y_1(t) + \beta y_2(t)$를 만족할 때 superposition이라 한다.



- Time invariant (시불변)

    system에 $x_1(t-t_0)$를 넣었을 때 $y_1(t-t_0)$가 나오는 경우

`-` Unit impulse function 단위 임펄스 함수



$$\delta[n]= 
\begin{cases}
{1,\quad\text{if n=0}}\\
{0,\quad\text{otherwise}}
\end{cases}$$

단위 임펄스 함수는 수에서 1과 같은 역할을 한다. 또, 벡터에서는 단위 기저벡터와 같은 역할을 한다.




`-` 이산 컨볼루션

단위 임펄스는 좌, 우로 평행이동이 가능하다. 가령 $\delta[n-2]$는 $\delta[n]$을 오른쪽으로 2만큼 shift 시킨 것

이 개념을 조금만 더 확장해보면 임펄스를 이용해 임의의 신호를 분해해 표현할 수 있게 된다. 아래의 임의의 이산 신호 $x[n]$을 보자.



이 이산신호는 $\delta[n]$ 및 이것을 좌우로 shift한 함수들을 이용해 다음과 같이 표현할 수 있다.

$$x[n]=\sum^\infty_{k=-\infty}x[k]\delta[n-k] \ \ \ \ (2)$$

$$= \dots + x[-2]\delta[n+2] + x[-1]\delta[n+1] + x[0]\delta[n] + x[1]\delta[n-1] + x[2]\delta[n-2] +\dots\ \ \ \ (3)$$

즉, $x[n]$이라는 임의의 함수를 $\delta[n-k]$라는 함수를 이용해서 분해할 수 있다.

![임의의 이산함수 x[n]를 단위 임펄스 함수를 이용해 분해한 결과](attachment:image-2.png)

식 (3)은 $x[n]$을 단위 임펄스 함수로 분해한 것으로 볼 수도 있고, 상수가 곱해진 신호열의 합으로 볼 수도 있다.

여기서 상수는 $x[k]$를 의미한다. $x[k]$는 $k$번째 함수값이라는 의미이기 때문에 더 이상 함수가 아닌 상수이다.

식 (3)은 다음과 같이 기호로 표기하기도 한다.

$$\Rightarrow x[n]*\delta[n]$$

이 때, $*$연산을 컨볼루션이라고 부르기도 한다.

일반적으로 임의의 이산 함수 $x[n]$과 $h[n]$에 대해 두 함수의 컨볼루션은 다음과 같다.

$$x[n]*h[n] = \sum^{\infty}_{k=-\infty}x[k]h[x-k]$$


`-` 이산 컨볼루션의 성질

- 교환 법칙

두 신호 $x[n]$과 $h[n]$에 대해 다음과 같이 교환 법칙이 성립

$$x[n]*h[n] = h[n]*x[n]$$

1. 입력 신호와 시스템의 임펄스 응답의 역할을 바꾸어도 시스템의 출력은 여전히 같다는 것을 의미
2. 두 개의 시스템이 직렬로 연결되는 경우 연결 순서를 바꾸어도 전체 시스템은 동일하다는 것

- 결합 법칙

컨볼루션은 셋 또는 그 이상의 신호에 대해 조합된 컨볼루션의 계산 순서에 영향을 받지 않는다.

$$(x[n]*h_1[n])*h_2[n] = x[n]*(h_1[n]*h_2[n])$$

예를 들어 두개의 시스템이 직렬로 연결되어 있는 경우, 두 개의 시스템을 컨볼루션으로 결합한 시스템 하나로 두 개 시스템의 결합을 대체할 수 있다는 것을 의미.

- 분배 법칙

$$x[n]*(h_1[n]+h_2[n]) = x[n]*h_1[n]+x[n]*h_2[n]$$

분배 법칙은 두 개의 시스템이 병렬로 연결되어 있을 경우 각 시스템의 임펄스 응답을 더한 것으로 전체 시스템을 대체할 수 있다는 것을 의미.

## 임펄스 응답(impulse response)

### memoryless system
- 현재만 영향을 받는다.

### causal system
- 과거 혹은 현재만 영향을 받는다.

### stable (BIBO stable) system
- $|x(t)|<\infty \quad \rightarrow \quad \square \quad \rightarrow \quad |g(t)|<\infty$ 
- 수렴하는 인풋이 시스템에 들어갔을 때 그 아웃풋 역시 발산하지 않는다.
- 발산하는 인풋을 넣었을 때는 그 아웃풋이 발산하는 발산하지 않든 BIBO라 이야기 할 수 없다.

    어떤 bounded 된 input에 대해서든 bounded 된 아웃풋이 나와야 BIBO라 할 수 있다.

- $|y(t)| = |\int x(\tau)h(t-\tau)d\tau| \le \int{\color{red}|x(\tau)h(t-\tau)|}d\tau = \int{\color{red}|x(\tau)||h(t-\tau)|}d\tau$ $<B\int^\infty_{-\infty}|h(t-\tau)|d\tau$

    - 적분의 절대값보다 절대값의 적분이 더 크거나 같다.

    - 모든 복소수에 대하여 

        $|A_1e^{j\theta_1}A_2e^{j\theta_2}| = |A_1A_2e^{j(\theta_1+\theta_2)}| = |A_1A_2e^{j(\theta_1+\theta_2)}| = A_1A_2$

        $|A_1e^{j\theta_1}||A_2e^{j\theta_2}| = A_1A_2$

        $\therefore$ $|xy| = |x||y|$

    - $\int^\infty_{-\infty}|h(t)|dt<\infty$ ${\color{yellow}\quad \Leftrightarrow \quad BIBO}$
